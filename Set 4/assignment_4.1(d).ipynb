### _Q.1(d)_

Create a plot to visualise what your program is doing?

```python
N = 10000
```

```python
def MCPi(N):
    count = 0
    np.random.seed(N)
    x_in  = np.array([], dtype = np.float64 )
    x_out = np.array([], dtype = np.float64 )
    y_in  = np.array([], dtype = np.float64 )
    y_out = np.array([], dtype = np.float64 )
    
    for i in range(N):
        x = np.random.uniform ( -1 ,1)
        y = np.random.uniform ( -1 ,1)
        r2 = x**2 + y**2
        dist = np.sqrt(r2)
        
        if dist <= 1:
            count = count + 1
            x_in = np.append(x_in, x)
            y_in = np.append(y_in, y)
        
        else :
            x_out = np.append(x_out, x)
            y_out = np.append(y_out, y)

    pi = 4*count/N
    
    # using patches
    rect = plt.Rectangle(( -1 , -1), 2, 2, fc='white', ec='red', fill=False )
    plt.gca ().add_patch(rect)
    circ = plt.Circle((0, 0), 1.0, fc = 'yellow', ec ='red', alpha =0.3)
    plt.gca().add_patch(circ)
    
    plt.axis('scaled')
    plt.scatter(x_in, y_in , s=0.4 , c='g')
    plt.scatter(x_out, y_out , s=0.4 , c='b')
    plt.title(r' Trials = {} , $ \ pi$ = {}'.format(N , pi))
    
    return pi
```

```python
MCPi(1000)
```
